```python
import pandas as pd
```

```python
df = pd.read_csv('../data/raw_data.csv', sep = '#')
df
```

```python
df['Year']= [i[0] for i in df['Date'].str.split('-',2)]

df
```

```python
df = df[['Year','Genre']]



limpa = lambda x: x[2:-1].split("\'",1)[0]

items = [limpa(str(i)) for i in (df.Genre)]
items
df.Genre = df.Genre.apply(lambda row: limpa(str(row)))
```
