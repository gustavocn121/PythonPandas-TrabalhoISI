```python
import pandas as pd
```

```python
df = pd.read_csv('../data/raw_data.csv', sep = '#')
df.head(5)
```

```python

df = df[df.Genre.str.contains("500") == False]
df = df[df.Genre.str.contains("Sorry") == False]
df
```

```python
df['Year']= [i[0] for i in df['Date'].str.split('-',2)]


df
```

```python
df['Month']= [i[1] for i in df['Date'].str.split('-',2)]
```

```python
df['Day']= [i[2] for i in df['Date'].str.split('-',2)]
df
```

```python
limpa = lambda x: x[2:-1].split("\'",1)[0]


df.Genre = df.Genre.apply(lambda row: limpa(str(row)))
```

```python

df['Genre']
```

```python

df = df.drop('Date', axis='columns')
```

```python
df = df.drop('Position', axis='columns')
```

```python
df.to_csv('../data/data.csv')
df
```
